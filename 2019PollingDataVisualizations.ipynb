```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = pd.read_csv('2018_2019_Polling_Data.csv')
```

```python
#Question 1 a)
dates = data.Date.unique()
plt.figure(figsize=(10,10))
cpcMean = []
lpcMean = []
ndpMean = []
gpcMean = []
ppcMean = []
bqMean = []
bestDates = dates[::-1]
for i in range (len(bestDates)):
    yearData = data[data.Date == bestDates[i]]
    cpcMean.append(round(yearData.CPC.mean(), 2))
    lpcMean.append(round(yearData.LPC.mean(), 2))
    ndpMean.append(round(yearData.NDP.mean(), 2))
    gpcMean.append(round(yearData.GPC.mean(), 2))
    ppcMean.append(round(yearData.PPC.mean(), 2))
    bqMean.append(round(yearData.BQ.mean(), 2))
bestCPC = cpcMean[::-1]
bestLPC = lpcMean[::-1]
bestNDP = ndpMean[::-1]
bestGPC = gpcMean[::-1]
bestPPC = ppcMean[::-1]
bestBQ = bqMean[::-1]
plt.plot(bestDates, bestCPC, color='blue')
plt.plot(bestDates, bestLPC, color='red')
plt.plot(bestDates, bestNDP, color='orange')
plt.plot(bestDates, bestGPC, color='green')
plt.plot(bestDates, bestPPC, color='yellow')
plt.plot(bestDates, bestBQ, color='brown')
plt.xlabel('Polling Dates')
plt.ylabel('% of Popular Vote')
plt.xticks(rotation='vertical', fontsize=5)
plt.title('Data Visualization of Canadian Political Parties (2018-2019)')
plt.show()
```

```python
#Question 1 b)
sampleTotal = data.Sample.sum()
pollingFirms = data.PollingFirm.unique()
values = []
for i in range(len(pollingFirms)):
    firmData = data[data.PollingFirm == pollingFirms[i]]
    values.append(round((firmData.Sample.sum() / sampleTotal) * 100, 2))
fig1, ax1 = plt.subplots()
ax1.pie(values, explode=None, labels=pollingFirms, autopct='%1.1f%%', shadow=True, startangle=90, radius=5000)
ax1.axis('equal') 
plt.tight_layout()
plt.show()
```
